## Matrix decomposition Multiplicative iteration

```python
import numpy as np
m = 10
n = 5
V = np.random.randint(10,size = (m,n)).astype(float)
error_list = []
print(V)
for k in range(3,4):
    W = np.random.randint(5, size = (m,k)).astype(float)
    H = np.random.randint(5, size = (k,n)).astype(float)
    error = np.sum((V-W@H)**2)
    print(error)
    errors = [error]
    for i in range(50):
            H *= ((W.T@V) / (W.T@W@H))
            W *= ((V@H.T) / (W@H@H.T))
            print("Dimension: % d, Step: %d" % (k,i))
            print('H:')
            print(H)
            print('W:')
            print(W)
            print('diff:')
            print(V-W@H)
            error = np.sum((V-W@H)**2)
            errors.append(error)
    error_list.append(errors)
```
